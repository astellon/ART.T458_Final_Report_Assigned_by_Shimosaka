```python
def cost(x, y, w, λ):
    reg = λ * np.dot(w,w)
    costs = 0
    for i in range(len(y)):
        costs += max(0, 1 - y[i] * np.dot(w, x[i,:]))
    return costs + reg
```

```python
def negdual(x, y, alpha):
    K = np.matmul(y, y.T) * np.matmul(x, x.T)
    return - np.matmul(alpha.T, np.matmul(K, alpha)) / (4*λ) + np.sum(alpha)
```

```python
def project(x):
    D = len(x)
    
    if np.all(0 <= x) and np.all(x <= 1):
        return x
    
    return np.clip(x, 0, 1)
```

```python
def alpha2w(x, y, alpha, λ):
    D = x.shape[1]
    w = np.sum(np.tile(alpha.reshape(-1,1),(1,D)) * np.tile(y.reshape(-1,1),(1,D)) * x, axis=0)
    return 2 / λ * w
```

```python
def update(x, y, alpha, lr):
    K = np.matmul(y, y.T) * np.matmul(x, x.T)
    return project(alpha + lr * ( - np.matmul(K, alpha)/2/λ + np.ones(len(alpha))))
```

```python
x, y = dataset.two(200)
```

```python
# initial value
epochs = 1000
alpha = np.random.rand(200)
λ = 100
lr = 1e-5
```

```python
for i in range(epochs):
    alpha = update(x, y, alpha, lr)
    cost_lag = cost(x, y, alpha2w(x, y, alpha, λ), λ)
    cost_neg = negdual(x, y, alpha)
    print(f'[{i+1:>2}/{epochs:>2}] Cost: {cost_lag:5.2f}, Cost of Negative Dual Problem: {cost_neg:5.2f}, diff: {cost_lag - cost_neg:5.2f}')
```
